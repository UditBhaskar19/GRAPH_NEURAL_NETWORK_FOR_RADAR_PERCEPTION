```python
import numpy as np
import matplotlib.pyplot as plt

def plot_scatter(samples_xy, figsize):
    _, ax = plt.subplots(figsize=figsize)
    ax.scatter(samples_xy[:,0], samples_xy[:,1], 1, color='blue', marker='.', label='samples')
    ax.set_aspect('equal')
    plt.tight_layout()
    plt.show()

def compute_sample_mean_covariance(samples_xy):
    mu = np.sum(samples_xy, axis=0) / samples_xy.shape[0]
    error = np.expand_dims((mu[:2] - samples_xy[:, :2]), axis=-1)
    sigma = error @ error.transpose(0, 2, 1)
    sigma = np.sum(sigma, axis=0) / ( samples_xy.shape[0] - 1 )
    return mu, sigma

mu = np.array([0, 0], dtype=np.float32)
sigma = np.array([[4, -1],[-1, 1]], dtype=np.float32)
samples_xy = np.random.multivariate_normal(mean=mu, cov=sigma, size=10000)
sample_mu, sample_sigma = compute_sample_mean_covariance(samples_xy)

plot_scatter(samples_xy, figsize=(8,8))
plot_cov(samples_xy, sample_mu, sample_sigma, n=3)
```
